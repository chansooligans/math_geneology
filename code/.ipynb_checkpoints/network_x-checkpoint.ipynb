```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import string
import datetime
```

```python
df = pd.read_csv('../data/math_geneology_final.csv')
#df = df.head(500)
df = df.fillna(' ')
```

```python
df = df.query('student_school == "Columbia University" or advisor_school == "Columbia University"')
```

Nodes

```python
cols = ['id','name','school','year']
students = df[['student_id','student_name','student_school','student_year']]
advisors = df[['id','advisor_name','advisor_school','advisor_year']]
students.columns = cols
advisors.columns = cols
df_nodes = pd.concat([students,advisors],axis=0)
df_nodes = df_nodes.drop_duplicates().reset_index(drop=True)
```

```python
ids = df_nodes['id']
nodes_dict = [{'name':df_nodes.loc[i,'name'], 'school':df_nodes.loc[i,'school'], 'year':df_nodes.loc[i,'year']} for i in range(df_nodes.shape[0])]
nodes = [x for x in zip(ids,nodes_dict)]
```

Edges

```python
edges = [x for x in zip(df['id'],df['student_id'])]
```

# Manual Edits

```python
# edit 'Rold\x13an-Correa, Alejandro'
edit_index = [i for i, x in enumerate(nodes) if x[0] == 204647]
if edit_index:
    nodes[edit_index][1]['name'] = "".join(filter(lambda x: x in string.printable, nodes[edit_index][1]['name']))
```

# Add to Graph

```python
G = nx.Graph()
```

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

# Export Graph

```python
nx.write_graphml(G,'../data/g.xml')
```

# Plot Graph using Matplotlib

```python
degrees = [G.nodes[x]['school'] for x in G]
cat_to_float = dict([x for x in zip(set(degrees), range(len(degrees)))])
node_colors = [cat_to_float[degree] for degree in degrees] 
```

```python
figure(num=None, figsize=(64, 24), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(121)

nx.draw_networkx(G, with_labels=True, node_color = node_colors)

plt.show()
```
