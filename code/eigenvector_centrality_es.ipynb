```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import string
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digraph = nx.DiGraph()

nodes = pd.DataFrame({'nodes':['0','1','2','3','4','5','6','7']})
digraph.add_nodes_from(nodes.nodes)

edges = pd.DataFrame([
    ['7','0',1],
    ['7','1',1],
    ['7','2',1],
    ['7','3',1],
    ['5','4',1],
    ['5','6',1],
    ['5','7',1],
])
edges.columns = ['from', 'to', 'weight']
digraph.add_edges_from([(e['from'], e['to'], {'weight': e['weight']}) for e in edges.to_dict(orient="records")])
```

```python
nx.draw_circular(digraph)
```
